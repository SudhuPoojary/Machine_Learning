```python
#droping User ID column as it is not an important feature
DF.drop("User ID", axis = 1, inplace = True)
```

```python
DF.head()
```

```python
#converting categorical values to numerical
DF = pd.get_dummies(DF, drop_first = True)
```

```python
DF.head()
```

```python
#seperating X and y
X = DF.loc[:,['Age','EstimatedSalary','Gender_Male']].values
y = DF.loc[:,['Purchased']].values
```

```python
X
```

```python
y
```

```python
#Feature Scaling
#Whenever we are dealing with KNN algorithm, we should do the feature scaling
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X = sc.fit_transform(X)
```

```python
X
```

```python
#model was expecting a 1d array as input
y = y.reshape(-1)
```

```python
y
```

```python
y.shape
```

```python
#Splitting the data into train and test data
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,random_state = 0, test_size = 0.25)
```

```python
#importing classification metrics
from sklearn.metrics import confusion_matrix,accuracy_score
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
#Selecting the k value.
acc_list = []
err_list = []
for i in range(1,25):
    model = KNeighborsClassifier(n_neighbors=i)
    model.fit(X_train,y_train)
    y_pred = model.predict(X_test)
    acc = accuracy_score(y_test,y_pred)
#     print("For k = {}, accuracy = {}".format(i,acc))
    acc_list.append(acc)
    err_list.append(1-acc)
```

```python
plt.plot(list(range(1,25)),err_list,c= 'g')
plt.title('Error rate v/s K')
plt.xlabel('K')
plt.ylabel('Error rate')
plt.show()
```

```python
#NOTE : from the above graph we can select k value as 5
```

```python
#Model Building
model = KNeighborsClassifier(n_neighbors=5)
model.fit(X_train,y_train)
y_pred = model.predict(X_test)
```

```python
#Model Evaluation
confusion_matrix(y_test,y_pred)
```

```python
accuracy_score(y_test,y_pred)
```
