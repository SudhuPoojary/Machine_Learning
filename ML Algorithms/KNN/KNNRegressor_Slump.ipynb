```python
x = df.iloc[:,:-1].values
y = df.iloc[:,-1].values
```

```python
y = y.reshape(-1)
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X = sc.fit_transform(x)
```

```python
X
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_tr,X_ts,Y_tr,Y_ts = train_test_split(X,y, random_state = 0, test_size = 0.2)
```

```python
from sklearn.neighbors import KNeighborsRegressor
from sklearn.metrics import mean_squared_error, r2_score
```

```python
mse = []
r2 = []
for k in range(1,10):
    model = KNeighborsRegressor(n_neighbors=k)
    model.fit(X_tr, Y_tr)
    Y_prediction = model.predict(X_ts)
    mse_i = mean_squared_error(Y_ts, Y_prediction)
    mse.append(mse_i)
    r2_i = r2_score(Y_ts, Y_prediction)
    r2.append(r2_i)
    
```

```python
plt.figure(figsize = (8,8))
plt.plot(np.range(1,10),mse,'r')
plt.xlabel('k-value')
plt.ylabel('mean-squared-error')
plt.title('Selecting k-value')
plt.show()
```

```python
plt.figure(figsize = (8,8))
plt.plot(np.arange(1,10),r2,'g')
plt.xlabel('k-value')
plt.ylabel('mean-squared-error')
plt.title('Selecting k-value')
plt.show()
```
