### About Dataset

###### This dataset is a collection of 30000 women fashion products. Categories covered in this dataset is western wear, Indian wear, perfumes and fragrances, watches and nightwear. ✨

### About File:
###### BrandName: Mentions the brand of the product
###### Details: Deatils about the product
###### Size: Sizes available
###### MRP: This is max retail price
###### SellPrice: This is the price after discount
###### Category: Category of the product
###### Nan value is null value

```python
FashionData["Category"]=FashionData["Category"].str.replace("-Women", "")
```

```python
FashionData["Sizes"]=FashionData["Sizes"].str.replace("Size:", "")
```

##### Here we can say that Max MRP is 16999.00 and Max SP is 13599.00 and Min MRP is 171.00 and min SP is 114.00.

```python
FashionData["Discount"].unique()
```

```python
FashionData["Discount"]= FashionData["Discount"].str.replace("% off","")
FashionData["Discount"]
```

```python
FashionData["Discount"]=FashionData["Discount"].astype(int)
```

```python
def discount_range(i):
        if i in range(0,26):
            return str("0-25%")
        elif i in range(26,51):
            return str("26-50%")
        elif i in range(51,76):
            return str("51-75%")
        else:
            return str(">75%")
FashionData["Discount_range"] = FashionData["Discount"].apply(discount_range)
FashionData.head()
```

```python
FashionData.rename(columns={"Discount":"Discount_in_%"},inplace=True)
```

```python
FashionData.head()
```

#### MRP vs SELL PRICE

```python
plt.figure(figsize = (20,8))
sns.jointplot(x = FashionData["MRP"], y = FashionData["SellPrice"],kind = "scatter")
```

```python
plt.figure(figsize = (20,8))
sns.jointplot(x = FashionData["MRP"], y = FashionData["SellPrice"],kind = "reg")
```

There is a linear relation between MRP and SP.

```python
plt.figure(figsize = (20,8))
sns.jointplot(x = FashionData["MRP"], y = FashionData["SellPrice"],kind = "scatter",hue = FashionData["Discount_range"])
```

```python
sns.barplot(x=FashionData["Discount_range"], y= FashionData["MRP"])
```

Higher the MRP higher the discount, a product with MRP 4000 and discount 50%-75% or above 75% attracts more than a product with MRP 2000 and discount 30% or 25%. Consumers perceive a higher level of savings for a product when a higher price discount is provided. This perception plays a crucial role in price-discount-value relation.
